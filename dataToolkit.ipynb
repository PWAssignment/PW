### Q1 : Creating the identical 2D Array

```python
'''[[1, 2, 3],
    [1, 2, 3],
    [1, 2, 3]]'''
## Method 1  : Numpy tile
array1 = np.tile([1, 2, 3], (3, 1))
print('Array 1\n')
print(array1)

## Method 2 : List Multiplication
array2 = np.array([[1, 2, 3]] * 3)
print('\nArray 2\n')
print(array2)

## Method 3 : 
row = np.array([1, 2, 3])
array3 = np.ones((3, 1), dtype=int) * row
print('\nArray 3\n')
print(array3)
```

# Q3  
### Difference Between `np.array`, `np.asarray`, and `np.asanyarray`

| Function         | Description                                                                                   | Copy Behavior               |
|------------------|-----------------------------------------------------------------------------------------------|-----------------------------|
| `np.array()`     | Always creates a new NumPy array from the input.                                              | Makes a copy (by default).  |
| `np.asarray()`   | Converts input to array only if itâ€™s not already an array.                                    | Avoids copy if unnecessary. |
| `np.asanyarray()`| Same as `np.asarray()`, but preserves subclasses (e.g., `matrix`, `masked_array`).           | Avoids copy and preserves subclass. |

### Differece Between the Shallow copy and Deep Copy

| Type          | Definition                                                        | Memory Sharing |
|---------------|-------------------------------------------------------------------|----------------|
| Shallow Copy  | Creates a new object, but references the original data inside     | Shared         |
| Deep Copy     | Creates a new object and copies the data too                      | Independent    |

```python
arr = np.array([1, 2, 3])
shallow = arr.view()
deep = arr.copy()

arr[0] = 100 # chaning in the original array

print("Original:", arr)
print("Shallow:", shallow)  # changes happen
print("Deep:", deep)        # changes will not here
```

```python
import re
def clean_phone(phone):
    # 1) Handle true NaNs
    if pd.isna(phone):
        return pd.NA

    # 2) Work with a stripped string
    phone_str = str(phone).strip()

    # 3) Drop anything after an 'x' or 'X' (extension)
    phone_str = re.split(r'[xX]', phone_str)[0]

    # 4) Remove all non-digits
    digits = re.sub(r'\D+', '', phone_str)

    # 5) If nothing remains, mark as NA, else convert to int
    return pd.NA if digits == '' else int(digits)

# Apply and cast to pandas nullable Int64
df['Phone'] = df['Phone'].apply(clean_phone).astype('Int64')

print(df['Phone'].dtype)  
print(df['Phone'].head())
```

### Q14

```python
# Sample time-series data
np.random.seed(24) 
date_range = pd.date_range(start='2024-01-01', periods=30)
temperature = np.random.randint(20, 40, size=30)
humidity = np.random.randint(30, 90, size=30)

df_time = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

# Plot
fig, ax1 = plt.subplots(figsize=(10, 6))

# Temperature plot
line1, = ax1.plot(df_time['Date'], df_time['Temperature'], 'r-', label='Temperature')
ax1.set_ylabel('Temperature (Â°C)', color='red')

# Create second y-axis for humidity
ax2 = ax1.twinx()
line2, = ax2.plot(df_time['Date'], df_time['Humidity'], 'b--', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')

# Combine legends from both axes
lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left')

# Labels and title
ax1.set_xlabel('Date')
plt.title('Temperature and Humidity Over Time')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import norm

np.random.seed(45)
# Generate normal distribution data
data = np.random.normal(loc=0, scale=1, size=1000)

# Histogram
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# PDF overlay
pdf = norm.pdf(bins, 0, 1)
plt.plot(bins, pdf, 'k', linewidth=2, label='Normal PDF')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.grid(True)
plt.show()
```

### Q17

```python
np.random.seed(23)
x = np.random.uniform(-1, 1, 200)
y = np.random.uniform(-1, 1, 200)

# Determine quadrant
def quadrant(x, y):
    if x > 0 and y > 0:
        return 'Q1'
    elif x < 0 and y > 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'

df_quadrant = pd.DataFrame({'x': x, 'y': y})
df_quadrant['Quadrant'] = [quadrant(a, b) for a, b in zip(x, y)]

# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_quadrant, x='x', y='y', hue='Quadrant', palette='Set1')
plt.axhline(0, color='black', linestyle='--')
plt.axvline(0, color='black', linestyle='--')
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

### Q19

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource, HoverTool

output_notebook()
np.random.seed(46)

# Generate data
categories = [f'Cat {i}' for i in range(1, 6)]
values = np.random.randint(10, 100, size=5)

source = ColumnDataSource(data=dict(categories=categories, values=values, color=["#%06x" % np.random.randint(0, 0xFFFFFF) for _ in range(5)]))

# Plot
p = figure(x_range=categories, title="Random Categorical Bar Chart", x_axis_label='Category', y_axis_label='Value', width=700, height=400)
p.vbar(x='categories', top='values', width=0.5, color='color', source=source)

# Hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

show(p)
```
