# Plotly Express

### Q1

```python
np.random.seed(30)
data = {
    'x': np.random.uniform(-10, 10, 300),
    'y': np.random.uniform(-10, 10, 300),
    'z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
fig = px.scatter_3d(df, x='x', y='y', z='z', title='3D Scatter Plot')
fig.show()
```

### Q2

```python
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
fig = px.violin(df, x='Grade', y='Score', box=True, points="all", title="Scores by Grade")
fig.show()
```

### Q3

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')
fig = px.imshow(pivot, labels=dict(x="Month", y="Day", color="Sales"), title="Sales Heatmap")
fig.show()
```

### Q4

```python
import numpy as np
import pandas as pd
import plotly.express as px

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

df = pd.DataFrame({
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
})

fig = px.density_heatmap(df, x='X', y='Y', z='Z', histfunc="avg", nbinsx=50, nbinsy=50,
                         title='3D Surface Heatmap of z = sin(sqrt(x^2 + y^2))')
fig.show()
```

# Bokeh

### Q1

```python
from bokeh.plotting import figure, show, output_notebook

output_notebook()

x = np.linspace(0, 10, 500)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, line_width=2, line_color="blue")

show(p)
```

### Q2

```python
from bokeh.models import ColumnDataSource
import random

x = np.random.rand(100) * 10
y = np.random.rand(100) * 10
sizes = np.random.randint(10, 40, 100)
colors = [f"#{random.randint(0, 0xFFFFFF):06x}" for _ in range(100)]

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Random Scatter with Sizes and Colors")
p.scatter('x', 'y', size='sizes', color='colors', source=source)

show(p)
```

### Q3

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", y_axis_label='Count')
p.vbar(x='fruits', top='counts', width=0.5, source=source, color="orange")

show(p)
```

### Q4

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white")

show(p)
```
